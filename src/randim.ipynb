```python
def walk(nx,pos=0,step=1):
    points = []
    desl1  = []
    desl2  = []
    variac = []
    pos    = 0
    pos2   = 0
    step   = 1
    for i in range (0,nx):
        p = (randint(0,1))
        if p == 0 :
            pos  = pos - step
        else:
            pos = pos + step
        d1 = pos/(i+1)
        d2 = (pos2)/(i+1)
        v  = d2 - (d1**2)
        
        points.append(pos)
        desl1.append(d1)
        desl2.append(d2)
        variac.append(v)       
        
    return points,desl1,desl2,variac
```

```python
e1, d1,d2,v1 = walk(10)
```

```python
n
```

```python
ensemble_1,e_1_d1,e_1_d2,v1  = walk(n-1)
ensemble_2,e_2_d1,e_2_d2,v2  = walk(n-1)
ensemble_3,e_3_d1,e_3_d2,v3  = walk(n-1)
ensemble_4,e_4_d1,e_4_d2,v4  = walk(n-1)
ensemble_5,e_5_d1,e_5_d2,v5  = walk(n-1)
```

```python
index = []
for i in range(1,n):
    index.append(i)    
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
df['index']                = index
df['ensemble_1']           = ensemble_1 
df['ensemble_1_desloc']    = e_1_d1
df['ensemble_1_des_quad']  = e_1_d2
df['ensemble_1_variancia'] = v1

df['ensemble_2']           = ensemble_2
df['ensemble_2_desloc']    = e_2_d1
df['ensemble_2_des_quad']  = e_2_d2
df['ensemble_2_variancia'] = v2

df['ensemble_3']           = ensemble_3
df['ensemble_3_desloc']    = e_3_d1
df['ensemble_3_des_quad']  = e_3_d2
df['ensemble_3_variancia'] = v3

df['ensemble_4']           = ensemble_4
df['ensemble_4_desloc']    = e_4_d1
df['ensemble_4_des_quad']  = e_4_d2
df['ensemble_4_variancia'] = v4

df['ensemble_5']           = ensemble_5
df['ensemble_5_desloc']    = e_5_d1
df['ensemble_5_des_quad']  = e_5_d2
df['ensemble_5_variancia'] = v5
```
